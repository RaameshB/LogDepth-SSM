```python
import jax.numpy as jnp
from flax import nnx
import jax.random as jrandom
```

```python
rngs = nnx.Rngs(1)
```

```python
N = 6
D = jrandom.uniform(rngs.params(), (N,))
L = jrandom.uniform(rngs.params(), (N,))[:, None]
R = jrandom.uniform(rngs.params(), (N,))[:, None]
D, L, R
```

```python
jnp.exp(L @ R.T)
```

```python
trace = jnp.sum(L * R)
trace
```

```python
print(jnp.eye(N) + (jnp.e ** trace - 1)/trace * L @ R.T)
```
